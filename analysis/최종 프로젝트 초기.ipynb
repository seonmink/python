```python
def get_content(driver):
   
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    
    #contents
    try :
        content = soup.select('div.C4VMK > span')[0].text
        content = unicodedata.normalize('NFC',content)
    except:
        content = ''
    
    #tag
    tags =  re.findall(r'#[^\s#,\\]+', content)
    date = soup.select('time.FH9sR.Nzb55')[0]['datetime'][:10]
    
    # like
    try:
        like = soup.select('div.Nm9Fw > a > span')[0].text
    except: 
        like = 0 
    
    # 장소 
    try:
        place = soup.select('div.JF9hh')[0].text
        place = unicodedata.normalize('NFC', place)
    except:
        place = ''
        
    # 정보저장 
    data = [content, date, like, place, tags]
    return data

def move_next(driver):
    right = driver.find_element_by_css_selector("div.l8mY4.feth3")
    right.click()
    time.sleep(3)
    

    
    
    
    
    
```

```python
# 사용자 지정함수 테스트 
data = get_content(driver)
data
```

```python
move_next(driver)
```

```python
# 크롤링
results = []
target = 200
for i in range(target):
    try:
        data = get_content(driver)
        results.append(data)
        move_next(driver)
    except:      
        time.sleep(2)
        move_next(driver)
print(results[0:199])
```

```python
import pandas as pd
results_df = pd.DataFrame(results)
results_df.columns = ['content', 'date', 'like', 'place', 'tags']
results_df.to_excel('C:\\Users\\Administrator\\Desktop\\seonmin\\pythonwork\\analysis\\data\\최종/crawling_동작구.xlsx',index=False)
```
