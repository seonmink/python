```python
df1 = pd.DataFrame({
    "고객번호":[1001, 1002, 1003, 1004, 1005, 1006, 1007],
    "이름":["둘리", "도우너", "또치", "길동", "희동", "마이콜", "영희"]
})
print(df1)
print("=================================================================")
df2 = pd.DataFrame({
    "고객번호":[1001, 1001, 1005, 1006, 1008, 1001],
    "금액":[10000, 20000, 15000, 5000, 100000, 30000]
})
print(df2)
```

```python
pd.merge(df1,df2) # 동일한 열이 있다면 자동으로 그 열을 기준으로 데이터를 병합해준다.
```

```python
pd.merge(df1,df2,how='left') #--> df1에 있는 것은 다
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer').sort_values(by='금액',ascending=False)
```
