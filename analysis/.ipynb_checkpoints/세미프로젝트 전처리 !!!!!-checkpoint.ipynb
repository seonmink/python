```python
## 첫행을 칼럼으로 변경 
df = df.rename(columns=df.iloc[0])
df = df.drop(df.index[0])
df
```

```python
## 합계열 삭제
del df['합계']
```

```python
df.head()
```

```python
# 결측치 제거
df=df.drop(index=715)
```

```python
df
```

```python
df=df.set_index(['시도','시군구','사고월','사고년도'])
```

```python
df1=df.T
```

```python
df1
```

```python
df1=df1.stack('시도')
```

```python
df1=df1.stack('시군구')
```

```python
df1=df1.stack('사고월')
```

```python
df1
```

```python
df1=df1.reset_index()
```

```python
df1.rename(columns={'level_0':'사고일'},inplace=True)
df1.columns.name=''
```

```python
df1
```

```python
df1=df1.set_index(['사고일','시도','시군구','사고월'])
df1
```

```python
df2=df1.swaplevel('사고일','시군구')
```

```python
df2=df2.swaplevel('사고일','사고월')
df2
```

```python
df2=df2.swaplevel('시군구','시도')
df2
```

```python
df2=df2.reset_index()
```

```python
df2
```

```python
df2.rename(columns={'사고월':'월'},inplace=True)
df2.rename(columns={'사고일':'일'},inplace=True)
df2.rename(columns={"(경상자수)":'경상자수'},inplace=True)
df2.rename(columns={'(부상신고자수)':'부상신고자수'},inplace=True)
df2.rename(columns={'(중상자수)':'중상자수'},inplace=True)
```

```python
df2
```

```python
df2.loc[df2['경상자수']== '-',df['경상자수']]= 0
```

```python
df2=df2.replace({'-':0})
```

```python
df2['경상자수']= pd.to_numeric(df['경상자수'])
```
