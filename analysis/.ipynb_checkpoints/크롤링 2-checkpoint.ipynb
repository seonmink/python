```python
def get_content(driver):
   
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    
    #name
    try :
        name = soup.select('span.btxt')[0].text
        name = unicodedata.normalize('NFC',name)
    except:
        name = ''
    
      
    # menu
    try:
        menu = soup.select('span.stxt')[0].text
        menu = unicodedata.normalize('NFC',menu)
    except: 
        menu = '' 
    
    # content 
    try:
        content = soup.select('span.ctxt')[0].text
        content = unicodedata.normalize('NFC',content)
    except:
        content = ''
        
    # 정보저장 
    data = [name, menu, content, adress]
    return data

def move_next(driver):
    next1 = driver.find_element_by_css_selector("div.list-more-btn.type03")
    next1.click()
    time.sleep(3)
```
