```python
# 날짜 --> datetime으로 바꾸기
all2['날짜'] = pd.to_datetime(all2['날짜'])
all2
```

```python
# 요일 파생변수 
all2['요일']= all2['날짜'].dt.dayofweek
all2['요일'].replace({0: '월요일', 1:'화요일', 2:'수요일', 3:'목요일', 4:'금요일', 5:'토요일', 6:'일요일'},inplace=True)
all2
```

```python
# 월 파생변수 넣기 
all2['월']=pd.DatetimeIndex(all2['날짜']).month
all2
```

```python
# 단순 상관 관계 살펴보기 --> 상관이 굉장히 낮음을 알 수 있음. 
all2.corr()
```
