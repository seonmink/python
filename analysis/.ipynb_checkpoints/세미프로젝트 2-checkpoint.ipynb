```python
cr = pd.read_excel('data/2014_19년 서울_경기 개별 교통사고 정보 (2).xlsx')
cr.head()
```

```python
# 서울시만 뽑기
cr1=cr[cr['발생지_시도']=='서울']
cr1
```

```python
#발생일 datetime으로 바꾸기 
cr1['발생일']=cr1['발생일'].astype(str)
cr1['발생일']=pd.to_datetime(cr1['발생일'])
cr1.info()
print(cr1.head())
```

```python
cr1['발생일']=cr1['발생일'].astype('str')
sp = cr1['발생일'].str.split('-',expand=True)
sp.head()
```

```python
sp.columns=['연','월','일']
sp.head()
```

```python
cr2=pd.concat([cr1,sp],axis=1)
cr2.head()
```

```python
cr3 = cr2[(cr2['연'] == '2018') | (cr2['연'] =='2019')]
cr3
```

```python
cr3.isnull().sum()
cr3[cr3["법정동명"].isnull()]
cr3=cr3.dropna(axis=0)
```

```python
cr3.isnull().sum()
```

```python
# 인덱스 재설정
cr3
del cr3['index']
del cr3['level_0']
```

```python
cr3.head()
```
