```python
# 인덱스를 명시적으로 붙여주면 뭐가 좋을까??

##### 인덱스를 자유롭게 지정 가능 
s = pd.Series([9904312,3448737,2890451,2466052], index=['서울','부산','인천','천안']) #이름을 바꾸거나, 다른 모양으로 쓰일 수 있음.
print(s)                                            # 이걸 인덱스로 쓸 수 있음. 

print('--------------------------------------------------------')
# value와 index를 따로 구하기 
print(s.values,type(s.values)) # , 가 없음 -> 배열
print(s.index)

print('--------------------------------------------------------')
# 인덱스에 이름 지정
s.index.name='도시'
print(s)

print('--------------------------------------------------------')

# Serise 전체의 이름 지정
s.name = '인구'
print(s)
```

```python
##### 백터화 연산 -> 시리즈 자체가 np 배열이기 때문에 벡터화 연산이 가능하다.

s1=s/1000000
print(s1)
```

```python
##### 인덱싱

print(s[1])
print(s[1],s['부산']) # --> 바꾼 인덱스 명으로도 가져올 수 있다. 
print(s[1],s['서울'])

print('--------------------------------------------------------')

print(s[1:3]) # 범위 지정 가능
print(s['부산':'천안'])

print('--------------------------------------------------------')

print(s[[0,3,1]]) # 필요한 것만 가져올 수 있음.
print(s[['서울','천안','부산']]) # 보다 직관적

print('--------------------------------------------------------')

 # fancy index 가능 

print(s[(s>250e4) & (s<500e4)])
print('--------------------------------------------------------')

print(s.부산,',',s.서울) # 이 인덱스의 데이터(값이 나옴)
```

```python
##### Serise와 dict

print('서울' in s)
print('대전' in s) # 없는 인덱스 입력시 false가 나옴.

print('--------------------------------------------------------')
# 딕셔너리 문법과 비슷 

print(list(s.items()))

print('--------------------------------------------------------')
#딕셔너리 처럼도 사용할 수 있음.
s2=pd.Series({'서울':9631482,'부산':3448737,'인천':2632035,'대전':1490158})
print(s2)
```

```python
##### 인덱스 기반 연산 
print(s)

print('--------------------------------------------------------')  

print(s2)

print('--------------------------------------------------------')  
result = s-s2  # 인덱스씨리 연산 매칭 되는 것끼리만 없으면 nan값으로 표현 
print(result)


print('--------------------------------------------------------')  

# 값으로만 계산 
print(s.values - s1.values)
```

```python
##### 결측치 제거 

print(result)
print(result.notnull())
print(result[result.notnull()])

print('--------------------------------------------------------') 

# s와 s2 데이터를 이용해서 인구 증가율 : (끝 연도 - 시작연도)/시작연도 *100
result2 = (s2-s)/ s*100
print(result2)

result2 = result2[result2.notnull()]
print(result2)
```

```python
#### 데이터의 수정, 삭제, 갱신

result2['부산']= 1.63 # 수정
print(result2)

# 갱신 - 추가
result2['대구'] = 1.41
print(result2)

# 삭제
del result2['서울']
print(result2)
```

#### (7). Indexer
+ 인덱싱을 해주는 애 라고 생각하면 됨. 

+ loc : 라벨값 기반의 2차원 인덱싱(결국 데이터 프레임)을 지원하는 인덱서
    - df.loc[행 인덱싱 값,열인덱싱 값]
    - df.loc[0] --> 하나의 값만 입력한다면 그에 해당되는 하나의 행만 뽑아 온다고 생각하면 된다. --> df라는 데이터프레임에서 인덱스 이름이 0인 행만 출력해서 가져와라는 의미 
+ iloc : 순서를 나타내는 정수 기반의 인덱서 
    - 
+ ix: 옛날에 쓰던거. 몰라도 됨  
+ at : 숫자 문자 둘다 접근, loc, 값을 하나만
+ iat: iloc, 값을 하나만 뽑아옴

##### 1) loc

```python
df = pd.DataFrame(np.arange(10,22).reshape(3,4),index=['a','b','c'],
                 columns=['A','B','C','D'])
df
```

```python
#### 행에 접근
print(df[:1])
print(df[:'a']) # 행하나만 가져와도 인덱싱을 해줬어야했음 
print('--------------------------------------------------------------')
print(df.loc['a']) # 시리즈로 만들어줌, 행부터 접근
print('--------------------------------------------------------------')
print(df.loc[df['A']>15])#기본 인덱서와 같이 쓸 수 있음
```

```python
##### 행과 열에 접근 
#b 행 B열에 있는 하나의 값을 추출 
# 기본 인덱스 
print(df['B']['b']) # 열부터 접근 
#print(dfp['B','b']) 불가

print(df.loc['b']['B'])
print(df.loc['b','B']) 

print('--------------------------------------------------------------')
print(df.loc['b':,'B':])
print(df.loc[['a','c'],['B','C']])

print('--------------------------------------------------------------')

#모든 행에 대해 첫번째 행에 있는 값이 11보다 작거나 같은 행의 칼럼 조회
df.loc[:, df.loc["a", :] <= 11]
```

```python
# 반드시레이블로만 접근을 해야함 
#print(df.loc[1:,1:]) 반드시 문자로만 접근해야함.
```

### iloc
    - 데이터 프레임의 행이나 칼럼의 순서를 나타내는 정수로 특정 값을 추출해오는 방법. loc는 칼럼명을 직접 적거나 특정 조건식을 써줌으로써 사람이 읽기 좋은 방법으로 데이터에 접근하는 방법이었다면,
    - iloc는 컴퓨터가 읽기 좋은 방법으로(숫자로) 데이터가 있는 위치(순서)에 접근한다고 생각하면 됨. 
 
- df.iloc['행 인덱스','열인덱스']
    - df.loc[0] --> 전체 데이터프레임에서 인덱스 이름이 0인 행만 추출 
    - df.iloc[0] --> 전체 데이터 프레임에서 0번째 행에 있는 값들만 추출해줘

```python
print(df['A']['a'])
# print(df[0]) # 이미 칼럼명이 문자로 되어 있어서 숫자로 못불러옴 

print('--------------------------------------------------------------')
#df.loc[0,0]
print(df.loc['a','A'])
print('--------------------------------------------------------------')

print(df.iloc[0,0])
print(df.iloc[:2,2])#0부터 1 행까지, 2열
print(df.iloc[0,-2:])   

print('--------------------------------------------------------------')
''''''
    B C
c   19 20
''''''
```

#### (5) apply()

+ 행이나 열 단위로 더 복잡한 처리를 하고자 할 때 사용
+ 인수로 행 또는 열을 받는 함수를 apply()의 인자로 넣으면 각 열(또는 행)을 반복하여 그 함수에 적용 

```python
df = pd.DataFrame({
    'A':[1,3,4,3,4],
    'B':[2,3,1,2,3],
    'C':[1,5,2,4,4]
})
df
```

```python
##### 각 열(행)별로 최대값에서 최소값을 뺀 값을 구하고자 한다.

def diff(x):
    return x.max() -x.min()
print(diff(df['A']))
print(diff(df['B']))
print(diff(df['C']))

print('===============================================================')
#한번에 처리할 수가 있음. 
print(df.apply(diff,axis=0))# 열 기준
print(df.apply(diff,axis=1))# 행 기준
```

```python
##### 람다 함수 -> 함수를 별도로 만들필요 없음, 이름이 없는 한번만 사용하기 위한 함수.
df.apply(lambda x:x.max()-x.min(),axis=0)
```

```python
df1 = pd.DataFrame({
    "고객번호":[1001, 1002, 1003, 1004, 1005, 1006, 1007],
    "이름":["둘리", "도우너", "또치", "길동", "희동", "마이콜", "영희"]
})
print(df1)
print("=================================================================")
df2 = pd.DataFrame({
    "고객번호":[1001, 1001, 1005, 1006, 1008, 1001],
    "금액":[10000, 20000, 15000, 5000, 100000, 30000]
})
print(df2)
```

```python
#알아서 inner join 이 됨. 간단하게 
pd.merge(df1,df2)
pd.merge(df1,df2,how='left') # 여기서 left는 df1 그래서 df1에 있는건 다 나옴
pd.merge(df1,df2,how='right')
pd.merge(df1,df2,how='outer')# full outer --> 양쪽에 있는건 다나옴.
```

```python
##### apply(), agg(), transform()

import seaborn as sns
iris= sns.load_dataset('iris')
iris.head()
```

```python
# 각 품종별로 가장 큰 값과 가장 작은값의 비율을 조회

def peek_to_peek_ratio(x):
    return x.max()/x.min()

##################################
# 품질이 3개니까  3번 입력해야하는데 한번만에 해결하는게 있음
g= iris.groupby('species')
print(g.apply(peek_to_peek_ratio)) # g의 각 품종별로 함수를 적용해줌

print('--------------------------------')
```

```python
g.agg(peek_to_peek_ratio)
```

```python
g.transform(peek_to_peek_ratio) # 값은 다 나오는데 요약을 해주지 못함, 모든 데이터를 풀어서(모든 품좀) 
#새로운 열을 추가할때 유리 
```

```python
#apply, transform 은 한개의 함수만 가능
# but agg는 여러개의 함수 가능
iris.groupby('species').agg([np.sum,np.mean,np.std,peek_to_peek_ratio])
```

```python
import seaborn as sns

tips=sns.load_dataset('tips')
print(tips.head())
print(tips.tail())
print(tips.describe())
print(tips.info())
```

```python
pd.date_range('2021-1-1','2021-1-31',freq='B')
```

```python
##### shift 연산 - 날짜를 이동해서 연산 

np.random.seed(0)

ts= pd.Series(np.random.randn(4),index=pd.date_range('2021-1-1',periods=4,freq='M'))
print(ts)

print('----------------------------------')

print(ts.shift(1)) # 값을 옮기는 것 
print(ts.shift(2))

print('----------------------------------')
print(ts.shift(-1)) # 역으로 값을 옮기는 것.
```

```python
##### 인덱스를 이동시키는 것 
print(ts.shift(1,freq='M'))
print('----------------------------------')

print(ts.shift(1,freq='W'))
```

```python
##### resampling: 시간간격을 다시 재조정
# 시간 구간이 작아지면 데이터양이 증가 -> 데이터 양을 증가시키기 위한 방법: upsampling
# 시간 구간이 커지면 데이터양이 감소: downsampling

ts=pd.Series(np.random.randn(100),index=pd.date_range('2021-1-1',periods=100,freq='D'))
ts.tail(10)

# 다운샘플링 : 데이터가 묶이기 때문에 집계 연산을 해서 대표값을 구해야 한다. 
print(ts.resample('W').mean())
print('----------------------------------')

print(ts.resample('M').mean())
print('----------------------------------')

ts=pd.Series(np.random.randn(60),index=pd.date_range('2012-1-1',periods=60,freq='T'))
ts 
# 1분 단위로 6개 다운 샘플링 
print(ts.resample('10T').sum())
```

```python
#####unsampling --> 없던 데이터를 만들어 내야함. 
# forward filling : 앞에서 나온 데이터를 그대로 사용
# backwardfilling : 뒤에서 나올 데이터를 앞에서 미리 사용 
print(ts.head())
print('------------------------------------------------')

# forward filling
print(ts.resample('30S').ffill().head(10)) #앞에 있는 데이터를 끌어다가 데이터를 생성하는 것 
print('------------------------------------------------')

#backwardfilling
print(ts.resample('30S').bfill().head(10)) # 뒤에 있는 데이터를 끌어다가 데이터를 생성하는 것!!
```
