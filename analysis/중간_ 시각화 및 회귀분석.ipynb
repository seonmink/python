```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings(action="ignore")

import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
plt.rc("font", family=font_name)

import matplotlib as mpl
mpl.rcParams["axes.unicode_minus"] = False

from datetime import datetime
import datetime as dt
```

## 1. 데이터 불러오기

```python
df = pd.read_csv('data/2018-2019 서울,인천,경기 교통사고 및 강수량 데이터.csv')
df
```

```python
del df['Unnamed: 0']
```

```python
df.info()
```

## 2. 날짜 datetime형식으로 바꾸기

```python
df['날짜'] = pd.to_datetime(df['날짜'])
df.info()
```

## 3.  '요일' 파생변수 넣기

```python
df["요일"] = df["날짜"].dt.dayofweek
df["요일"].replace({0: '월요일', 1:'화요일', 2:'수요일', 3:'목요일', 4:'금요일', 5:'토요일', 6:'일요일'},inplace=True)
df
```

## 3. 월 파생변수 넣기

```python
df["월"] = pd.DatetimeIndex(df['날짜']).month
df
```

## 6. 정규화 해보기

```python
columns = ["사망자수","중상자수","경상자수","부상자수","사고건수","강수량(mm)","1시간최다강수량(mm)"]
df_2 = df2[columns]

from sklearn import preprocessing

x = df_2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_scaled = pd.DataFrame(x_scaled)
df_scaled
```

```python
column = ["날짜","지점명","요일", "월"]
df_3 = df2[column]
df_3 = df_3.reset_index(drop=True)
df_3
```

```python
df3 = pd.concat([df_3,df_scaled],axis=1)
df3 = df3.rename(columns={0:"사망자수",1:"중상자수",2:"경상자수",3:"부상자수",4:"사고건수",5:"강수량(mm)",6:"1시간최다강수량(mm)"})
df3
```

```python
sns.heatmap(data = df3.corr(), annot=True, fmt = '.2f', linewidths=.5, cmap='Blues')
```

## 지역별 사망자수 / 사고건수

```python
ta = df.groupby('지점명').sum()[['사고건수','사망자수']]
ta
```

```python
ta['사망자비율'] = ta['사망자수'] / ta['사고건수']
```

```python
ta.sort_values(by='사망자비율',ascending=False)
```
