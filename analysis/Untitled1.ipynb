```python
#날짜 datetime으로 바꾸기 
df['날짜']=df['날짜'].astype(str)
df['날짜']=pd.to_datetime(df['날짜'])
df.info()
print(df.head())
```

```python
df['날짜']=df['날짜'].astype('str')
sp = df['날짜'].str.split('-',expand=True)
sp.head()
```

```python
sp.columns=['연','월','일']
sp.head()
```

```python
df2=pd.concat([df,sp],axis=1)
df2.head()
```

```python
df3=df2.groupby('일')[['강수량','1시간최다강수량','사망자수','중상자수','경상자수','부상신고자수','건수']].mean()
```
